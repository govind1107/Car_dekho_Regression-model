```python
X = df_1.drop(columns = 'selling_price')
```

```python
Y = df_1['selling_price']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X[['km_driven','mileage','engine','max_power']] = scaler.fit_transform(X[['km_driven','mileage','engine','max_power']])
```

```python
X.head()
```

```python
#X['seats'] = X['seats'].astype(str)
```

```python
X.info()
```

```python
X = pd.get_dummies(X, ['fuel','seller_type','transmission','owner', 'seats','company'])
```

```python
X.shape
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=123)
```

```python
xgb = xgb.XGBRegressor(n_estimators=1000, max_depth=7, eta=0.1, subsample=0.9, colsample_bytree=0.8)
```

```python
xgb.fit(X_train,Y_train)
```

```python
xgb.score(X_test,Y_test)
```

```python
y_pred = xgb.predict(X_test)
```

```python
mean_absolute_percentage_error(Y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
rms = sqrt(mean_squared_error(Y_test, y_pred))
```

```python
print(rms)
```

```python
rmse = mean_squared_error(Y_test, y_pred , squared=False)
```

```python
print(rmse)
```

```python
mean_absolute_error(Y_test, y_pred)
```
